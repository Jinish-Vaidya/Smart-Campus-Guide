```python
import tensorflow as tf
model=tf.keras.models.load_model("(10 Epoch)efficientnetb0.h5")
```

```python
	# Save the model.
with open('model_efficientnetb0.tflite', 'wb') as f:
  f.write(tflite_model)
```
